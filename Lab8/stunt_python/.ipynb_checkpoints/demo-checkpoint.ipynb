```python
ble = get_ble_controller()
ble.connect()
```

```python
import re

sensor_data_index = 0
sensor_data = [[0.0 for _ in range(4)] for _ in range(2500)]

msg = None

def data_collect(uuid, byte_array):
    global sensor_data_index
    msg = ble.bytearray_to_string(byte_array)
    data = re.findall(r'-?\d+\.\d+', msg)
    sensor_data[sensor_data_index] = [float(num) for num in data]
    sensor_data_index += 1

ble.start_notify(ble.uuid['RX_STRING'], data_collect)
```

```python
ble.send_command(CMD.FLIP, "8|280|5000|750|1.25")  # duration|max_pwm|stop_time_when_no_wall|target_distance
```

```python
ble.send_command(CMD.STOP, "")
```

```python
print(sensor_data_index)
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array(sensor_data)
time = data[:sensor_data_index, 0]
tof = data[:sensor_data_index, 1]
pwd = data[:sensor_data_index, 2]
kf = data[:sensor_data_index, 3]
```

```python
plt.figure()
plt.plot(time, tof, label='tof')
plt.plot(time, pwd, label='pwm')
# plt.plot(time, kf, label='kf')
plt.title('Distance&PWM vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Distance&PWM')
plt.legend()
plt.grid()
plt.show()
```
